# Problem 1: The Law of Large Numbers

1. Draw `10^4` values from a Bernoulli distribution with `p = 0.5` and make a plot of the running average `S_n`. (The running average `S_n` is the average of the first `n` numbers, where `1 ≤ n ≤ 10^4`.)

2. Make a histogram of the means drawn from the Bernoulli with `p = 0.5` and sample sizes of `n = 100` and `n = 1000`.  Overplot a Gaussian with the same mean and variance.

```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline

trials = 10**4  # number of values
p = 0.5         # probability of success
Sn = []
values = []

for i in range(trials):
    values.append(np.random.binomial(1, p))
    
for n in range(trials):
    Sn.append(np.sum(values[:n]) / float(n+1))

plt.plot(Sn)
plt.title(r'Running average of $S_n$')
plt.xlabel('$n$')
plt.xscale('log')
plt.ylabel('$S_n$')
```

```python
means = []
for i in range(300):
    means.append(np.mean(np.random.binomial(1, p, size = 100)))

c, bins, t = plt.hist(means, 50)

sigma = 0.25
mu = 0.5
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ))
```

```python
plt.hist(Sn[:1000], 50)
sigma = 0.25
mu = 0.5
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ))
```
